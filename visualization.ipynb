```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# matplotlib

- matplotlib의 두 가지 중요 요소
1. figure: 캔버스 역할, 그림판의 크기와 색상 조절, 그래프 최종적으로 화면에 표시
2. axes: 실제 그림을 그리는 메소드 보유

#### figure()의 facecolor: 배경색 설정

```python
plt.figure(figsize=(4,4), facecolor='yellow')
plt.plot([1,2,3], [2,4,6])
plt.title('Hello plot')
plt.show()
```

### subplots(): 1개의 figure객체와, 여러 axes객체 생성

```python
f, ax = plt.subplots(1, 2, figsize=(6,3))
plt.show()
```

### plot(): 선그래프

##### 선 색 변경

```python
plt.plot([1, 2, 3], [4, 5, 6], color='green')
plt.show()
```

##### 좌표점 형태와 크기

```python
plt.plot([1, 2, 3], [4, 5, 6], marker='o', markersize=12)
plt.show()
```

##### 선의 형태

```python
plt.plot([1,2,3], [4,5,6], linestyle='dashed')
plt.show()
```

##### 선 두께

```python
plt.plot([1,2,3], [4,5,6], linewidth=4)
plt.show()
```

### 축 이름 설정

```python
plt.plot([1, 2, 3], [4, 5, 6], color='green', marker='o', markersize=15, linewidth=5, linestyle='--')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

### 축 눈금값 회전

```python
plt.plot(np.arange(100), 2*np.arange(100), color='g')

plt.xticks(rotation=90) # 90도 회전

plt.show()
```

### 눈금값 간격

- 축 값이 문자열: 자동으로 모든 값이 축에 표현됨
- 축 값이 수치형: 최솟값과 최댓값에 기반해 축의 크기에 따라 눈금 간격 생성

```python
plt.plot(np.arange(100), 2*np.arange(100))
plt.xticks(ticks=np.arange(0,100,5), rotation=90)
plt.show()
```

### 범례 설정

```python
plt.plot(np.arange(100), 2*np.arange(100), color='green', label='green line')
plt.plot(np.arange(100), 4*np.arange(100), color='red', label='red line')
plt.legend()
plt.show()
```

### 선과 막대그래프 함께 그리기

```python
plt.plot(np.arange(10), 2*np.arange(10), color='red', marker='o', linestyle='dashed', label='red line')
plt.bar(np.arange(10), 2*np.arange(10), color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()

plt.show()
```

### 선과 막대그래프 함께 그리기 (axes 객체 사용)

```python
f = plt.figure()
ax = plt.axes()

ax.plot(np.arange(10), 2*np.arange(10), color='red', marker='o', linestyle='dashed', label='red line')
ax.bar(np.arange(10), 2*np.arange(10), color='green', label='bar plot')

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax.legend()

plt.show()
```

### subplots() 이용해 그래프 그리기

```python
f, ax = plt.subplots(2, 2, figsize=(6, 6))

ax[0, 0].plot(np.arange(1, 10), 2*np.arange(1, 10), color='red', marker='o', linestyle='dashed', label='red line')
ax[0, 1].bar(np.arange(1, 20), 2*np.arange(1, 20), color='green', label='bar plot')
ax[1, 0].plot(np.arange(1, 10), 2*np.arange(1, 10), color='green', marker='o', linestyle='dashed', label='red line')
ax[1, 1].bar(np.arange(1, 20), 2*np.arange(1, 20), color='red', label='bar plot')


ax[0, 0].set_xlabel('ax[0, 0] x axis')
ax[0, 1].set_xlabel('ax[0, 1] x axis')
ax[1, 0].set_xlabel('ax[1, 0] x axis')
ax[1, 1].set_xlabel('ax[1, 1] x axis')

ax[0, 0].legend()
ax[0, 1].legend()
ax[1, 0].legend()
ax[1, 1].legend()

plt.show()
```

# seaborn

```python
titanic_df = pd.read_csv('titanic_train.csv')
```

**정보에 종류에 따른 시각화 차트 유형**
- histogram: 연속형 값에 대한 도수 분포(건수)
- countplot: 이산형 값(범주형 값)에 대한 건수 시각화
- barplot: 특정 칼럼의 이산 값에 따른 다른 칼럼의 연속형 값
- boxplot: 연속형 값의 사분위수, 최대, 최소, 이상치 시각화
- violinplot: 히스토그램과 박스플롯을 바이올린 형태로 함께 시각화
- scatterplot: 2개의 연속형 값 시각화
- heatmap: 연속형 컬럼들의 상관관계 시각화

### 히스토그램(Historgram)
- histplot(): axes 레벨 함수
- displot(): figure 레벨 함수

##### histplot

```python
f = plt.figure(figsize=(8, 3))
sns.histplot(titanic_df['Age'], bins=20)
plt.show()
```

```python
f = plt.figure(figsize=(8,3))
sns.histplot(data=titanic_df, x='Age', bins=30, kde=True)
plt.show()
```

##### displot

```python
f = plt.figure(figsize=(8,4))
sns.displot(data=titanic_df, x='Age', kde=True)
plt.show()
```

- figsize=(8,4) 반영안됨  
=> figure 레벨 함수는 plt.figure()를 이용해 크기 조절 불가!  
       
  
- height와 aspect 통해 크기 조절  
=> height는 높이, aspect는 가로와 세로의 배율  
=> 가로: height, 세로: height * aspect

```python
sns.displot(data=titanic_df, x='Age', kde=True, height=4, aspect=2)
plt.show()
```

### countplot

```python
sns.countplot(data=titanic_df, x='Pclass')
plt.show()
```

### barplot

```python
sns.barplot(data=titanic_df, x='Pclass', y='Age')
plt.show()
```

- y축 값은 기본적으로 평균값 나타냄
- estimator 매개변수 수정하여 변경 가능

```python
sns.barplot(data=titanic_df, x='Pclass', y='Survived')
plt.show()
```

```python
sns.barplot(data=titanic_df, x='Pclass', y='Sex')
plt.show()
```

```python
sns.barplot(data=titanic_df, x='Pclass', y='Survived', estimator=sum)
plt.show()
```

- estimator 매개변수를 sum으로 변경 => 객실 등급별생존자 수 총합

##### hue: 2차원 그래프에서 3개 변수의 정보를 나타낼 수 있게 함

```python
sns.barplot(data=titanic_df, x='Pclass', y='Age', hue='Sex')
plt.show()
```

```python
sns.barplot(data=titanic_df, x='Pclass', y='Survived', hue='Sex')
plt.show()
```

### boxplot

- x축에 데이터 입력 시 수평 박스 플롯
- y축에 데이터 입력 시 수직 박스 플롯

```python
sns.boxplot(data=titanic_df, y='Age')
plt.show()
```

```python
sns.boxplot(data=titanic_df, x='Pclass', y='Age')
plt.show()
```

```python
sns.boxplot(data=titanic_df, x='Pclass', y='Age', hue='Sex')
plt.show()
```

### violinplot

- 히스토그램의 연속확률분포 곡선과 박스 플롯 함께 시각화
- x축에 데이터 입력 시 수평 바이올린 플롯
- y축에 데이터 입력 시 수직 바이올린 플롯

```python
sns.violinplot(data=titanic_df, y='Age')
plt.show()
```

```python
sns.violinplot(data=titanic_df, x='Pclass', y='Age')
plt.show()
```

```python
sns.violinplot(data=titanic_df, x='Pclass', y='Age', hue='Sex')
plt.show()
```

# pandas의 DataFrame과 numpy의 ndarray는 서로 변환 가능

# sklearn에서는 ndarray 개체 많이 사용

### list => dataframe

```python
pd.DataFrame([1, 2, 3], columns=['col1'])
```

### dict => dataframe

```python
pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
```

### ndarray => dataframe

```python
pd.DataFrame(np.arange(3), columns=['col1'])
```

### dataframe => list

```python
pd.DataFrame([[1, 2], [3,4]], columns=['col1', 'col2']).values.tolist()
```

### dataframe => dict

```python
pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]}).values.tolist()
```

### dataframe => ndarray

```python
pd.DataFrame(np.arange(5)).values
```

## agg

```python
df['age'].agg(['mean', 'std'])
```

```python
df.groupby('gender').agg({'age': 'max', 'age_o':'min'})
```

```python
df['age_band'] = df['age'].map(lambda x: '20s' if x< 30 else '30s' if x < 40 else 'old' )
```

```python
df.groupby('age_band')['age'].count()
```
